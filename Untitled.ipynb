```python
import pandas as pd   #             
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import time
#  
df = pd.read_csv('iris.csv')
print("dataset: \n", df)

#   ()   
df['variety'] = df['variety'].map({'Setosa': 1, 'Versicolor': 2, 'Virginica': 3})

# ()   ()  
x = df.drop('variety', axis=1).values
y = df['variety'].values

#       
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

#  
def h(x, theta):
    z = x.dot(theta)
    return 1 / (1 + np.exp(-np.clip(z, -500, 500)))

#  
def log_loss(y_pred, y_true):
    epsilon = 1e-10
    loss = -y_true * np.log(y_pred + epsilon) - (1 - y_true) * np.log(1 - y_pred + epsilon)
    return loss.mean()

#  
mu = x_train.mean(axis=0)
sigma = x_train.std(axis=0)
xn_train = (x_train - mu) / sigma
xn_test = (x_test - mu) / sigma

#         
xn_train = np.insert(xn_train, 0, 1, axis=1)
xn_test = np.insert(xn_test, 0, 1, axis=1)

#   
num_classes = len(np.unique(y))

#      
alpha = 0.01
theta = np.random.randn(num_classes, xn_train.shape[1])

start_time = time.time() ##############################      
#      
def update_step():
    global theta
    for cls in range(num_classes):
        y_train_cls = np.where(y_train == cls + 1, 1, 0)
        y_pred = h(xn_train, theta[cls,:])      
        dtheta = np.zeros([xn_train.shape[1]])
        dtheta[0] = (y_pred - y_train_cls).mean()
        for i in range(1, xn_train.shape[1]):
            dtheta[i] = (xn_train[:, i] * (y_pred - y_train_cls)).mean()
        theta[cls,:] -= alpha * dtheta
        
accuracy_list=[] 
costs_train = []  #   
costs_test = []   #   
for i in range(700):
    update_step()
    #     
    #train_loss = []
    #test_loss = []
    #for cls in range(num_classes):
        #y_train_cls = np.where(y_train == cls + 1, 1, 0)
        #y_test_cls = np.where(y_test == cls + 1, 1, 0)
        #train_loss.append(log_loss(h(xn_train, theta[cls,:]), y_train_cls))
        #test_loss.append(log_loss(h(xn_test, theta[cls,:]), y_test_cls))
    #costs_train.append(np.mean(train_loss))
    #costs_test.append(np.mean(test_loss))
    #      
    y_probs = np.zeros((len(xn_test), num_classes))
    for cls in range(num_classes):
        y_probs[:, cls] = h(xn_test, theta[cls,:])
    y_pred_test = np.argmax(y_probs, axis=1)
    #   
    accuracy = np.mean(y_pred_test + 1 == y_test)
    accuracy_list.append(accuracy)
print("Accuracy:", accuracy)
   
end_time = time.time() ####################################     
test_training_time = end_time - start_time ####     
print("test_training_time:", test_training_time)


#     
#plt.plot(costs_train, label='costs_train')
#plt.plot(costs_test, label='costs_test')
#plt.xlabel('Iteration')
#plt.ylabel('Loss')
#plt.legend()
#plt.show()


#    
plt.plot(accuracy_list, label='accuracy_list Set')
plt.xlabel('iteration')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


#              
plt.scatter(range(len(y_test)), y_pred_test + 1, label='y_pred ', color='blue', marker='o', alpha=0.5)
plt.scatter(range(len(y_test)), y_test, label='y_test', color='red', marker='o', alpha=0.5)
plt.xlabel('Data Index')
plt.ylabel('Labels')
plt.legend()
plt.show()

```
